```python
import os
import openpyxl
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import pandas as pd
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
```

```python
data = pd.read_excel("C:/Users/User/Documents/Operation Holy Grail/Phase 4//Data/Pork n beef import/MPSA/agg_count.xlsx")
data.head()
data.info()
```

```python
import matplotlib.font_manager
#[f.name for f in matplotlib.font_manager.fontManager.ttflist]
```

```python
#Arial#
fig, ax = plt.subplots(figsize=(20, 5))
ax.plot(data['date_full'],
        data['count_full'],
        color='black')

plt.axvline(pd.to_datetime('2012-03-01'),
            color='blue', linestyle='--', lw=1, label=r"President Ma$'$s second ban-lifting announcement")
plt.axvline(pd.to_datetime('2016-05-20'),
            color='red', linestyle='--', lw=1, label='Political alternation')
plt.axvline(pd.to_datetime('2020-08-28'),
            color='green', linestyle='--', lw=1, label=r"President Tsai$'$s ban-lifting announcement")

legend = ax.legend(loc='upper left', prop={'family': 'Arial'})
legend.get_frame().set_linewidth(0.5) 
#legend.get_frame().set_edgecolor('black')
legend.get_frame().set_facecolor('none')
legend.set_bbox_to_anchor((0.005, 0.99))

ax.set_xlabel("Date", fontname="Arial", fontsize=12)
ax.set_ylabel("Count", fontname="Arial", fontsize=12)
ax.set_xlim(["2009-09-30", "2022-06-01"])

for tick in ax.get_xticklabels():
    tick.set_fontname("Arial")
for tick in ax.get_yticklabels():
    tick.set_fontname("Arial")

#Setting the date format#
date_form = DateFormatter("%Y-%m")
ax.xaxis.set_major_formatter(date_form)

#Date tick (x-axis)#
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))

plt.show()
fig.savefig('speechcount_ts.png', dpi=300)
```
